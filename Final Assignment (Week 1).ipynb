# Capstone Project - Final Assignment: Part 1

# 1. The challenge

My client wants to open a small restaurant in Amsterdam the Netherlands. She wants to know what would be the ideal location for her restaurant.

## Understand the problems

There are a few things that will influence the choice of location:

### Question 1: What type of food to serve?

For the client, the restaurant is a form of an investment. She does not have any preference what kind of food to serve. For her, the most popular food in the neighborhood would be the preference.
This qustion can be answered, hopefuly, by using Foursquare API.

Another type of data can help to validate the choice. That is the birth place of the residence in the area. Our assumption is that people's preference on food has a strong connection to where they are from.
We need demographic information to answer this question.

### Question 2: How many similar restaurants already exist?

Obviously, if this new restaurant would be the only one, there will be more profit for my client. So, the number of existing similar restaurants in the neighborhood must be taken into consideration.
Again, this qustion can be answered, hopefuly, by using Foursquare API.

### Question 3: Who are the target customers and where do they live?

It is going to be a small restaurant (5, 6 tables) due to the limited investment. The main income would be take out and orders made online.
From past experience, my client knows that people live alone are more likely to buy takeout or use online food ordering system such as Uber Eats. They are the ideal target customers for her new restaurant. So, we will look for an area that with relatively high density of one-person household.
We need demographic information to answer this question.

## 2.2. Source of demographic data

The Central Bureau of Statistics of the Netherlands, CBS in short, hosts a large number of demographic data regarding who live and work in the Netherlands.<br>
In [this page](https://opendata.cbs.nl/statline/#/CBS/nl/dataset/84286NED/table?ts=1558337871247), you can choose whatever feature you need to solve the problem. The list of features is quite comprehensive. It is important, therefore, to define the goal and requirement of data carefully.


#### List of features from CBS that can be used to answer the question

* **Regional specifics** _(Regioaanduiding)_: This information can help me to link relative details to a specific area.
* **Household** _(Particulier huishouden)_: People who live alone are less likely to cook themselves.
* **Population density** _(Bevolkingsdichtheid)_: A more densely populated area means more customers for a restaurant.
* **Origin of birth** _(Personen met een migratieachtergrond)_: This information can help us to adjust the types of cooking styles we offer.
* ~~**Income per household** _(Inkomen van huishoudens)_: Since restaurants are usually quite expensive comparing to home-cooking. Only people who have sufficient income can afford to go to restaurants often.~~
    * **Unfortunately, this information is missing consistently from the database.**

#### Features deliberately excluded

* **Civil status**: Since we can distinguish one-person households, it's not necessary to understand why people live alone (single or divorced does not seem to link to food strategy directly).
* **Gender**: Reason, in the Netherlands, there isn't a big difference between men and women in terms of cooking themselves.
* **Type of house**: This has no direct correlation to people's choice of food.
